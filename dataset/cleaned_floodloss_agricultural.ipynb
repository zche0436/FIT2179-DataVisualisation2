```python
import pandas as pd
```

```python
df = pd.read_excel("SPECIAL REPORT ON IMPACT OF FLOODS IN MALAYSIA, 2025.xlsx",
                   sheet_name = "Jadual 3.1",
                   skiprows=8)
```

```python
df.head()
```

```python
category_cols = ['Paddy', 'Crops', 'Fisheries','Livestock','Fisherman','Agro-entrepreneur', 'High Value Commodity', 'Total Loss']
```

```python
cols_2024 = [1,3,6,9,12,15,18,21,24]
cols_2025= [1,4,7,10,13,16,19,22,25]
```

```python
df_2024 = df.iloc[:, cols_2024].copy()
df_2024.columns = ['State'] + category_cols
df_2024['Year'] = 2024
```

```python
df_2025 = df.iloc[:, cols_2025].copy()
df_2025.columns = ['State'] + category_cols
df_2025['Year'] = 2025
```

```python
df_2024.head()
df_2025.head()
```

```python
mapshaper_states = {
    'Johor': 'Johor', 
    'Kedah': 'Kedah', 
    'Kelantan': 'Kelantan', 
    'Melaka': 'Melaka', 
    'Negeri Sembilan': 'Negeri Sembilan', 
    'Pahang': 'Pahang', 
    'Pulau Pinang': 'Pulau Pinang', 
    'Perak': 'Perak', 
    'Perlis': 'Perlis', 
    'Selangor': 'Selangor', 
    'Terengganu': 'Terengganu', 
    'Sabah': 'Sabah', 
    'Sarawak': 'Sarawak', 
    'W.P. Kuala Lumpur': 'Kuala Lumpur', 
    'W.P. Putrajaya': 'Putrajaya', 
    'W.P. Labuan': 'Labuan'
}
```

```python
def clean_df(df):
    df = df.copy()
    df['State'] = df['State'].astype(str).str.strip().str.replace('\n', ' ')
    df['State'] = df['State'].map(mapshaper_states)
    df = df.dropna(subset=['State'])
    for col in category_cols:
        df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0.0).round(0).astype(int)
    return df
```

```python
df_2024 = clean_df(df_2024)
df_2025 = clean_df(df_2025)
```

```python
cleaned_df = pd.concat([df_2024, df_2025], ignore_index=True)
cleaned_df = cleaned_df[['State', 'Year'] + category_cols]
```

```python
cleaned_df['State_Year'] = cleaned_df['State'] + '_' + cleaned_df['Year'].astype(str)
```

```python
cleaned_df.to_csv("cleaned_floodloss_agricultural.csv", index=False)
```
