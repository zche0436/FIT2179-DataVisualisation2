```python
import pandas as pd
```

```python
df_pr = pd.read_excel("era5-x0.25_timeseries_rx1day,r20mm,pr_timeseries_monthly_1950-2024_mean_historical_era5_x0.25_mean.xlsx", sheet_name ="pr")
```

```python
df_r20mm = pd.read_excel("era5-x0.25_timeseries_rx1day,r20mm,pr_timeseries_monthly_1950-2024_mean_historical_era5_x0.25_mean.xlsx", sheet_name ="r20mm")
```

```python
years = ('2020', '2021', '2022', '2023', '2024')
```

```python
def extract(df, value_col):
    date_cols = [col for col in df.columns if str(col).startswith(years)]
    df_filtered = df[['code', 'name'] + date_cols]

    df_long = df_filtered.melt(
    id_vars=['code', 'name'],
    var_name='YearMonth',
    value_name = value_col)

    df_long['Year'] = df_long['YearMonth'].str[:4].astype(int)
    df_long['Month'] = df_long['YearMonth'].str[5:].astype(int)
    df_long[value_col] = df_long[value_col].round(2)

    return df_long[['YearMonth', 'Year', 'Month', value_col]].sort_values(['Year', 'Month']).reset_index(drop=True)
```

```python
df_pr = extract(df_pr, 'pr')
df_r20mm = extract(df_r20mm, 'r20mm')
```

```python
df_cleaned = df_pr.merge(df_r20mm, on=['YearMonth', 'Year', 'Month'])
```

```python
month = {1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',
               7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'}
df_cleaned['MonthName'] = df_cleaned['Month'].map(month)
```

```python
df_cleaned.to_csv('cleaned_rainfall_2020_2024.csv', index=False)
```
